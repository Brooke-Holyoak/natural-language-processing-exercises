### Metacharacters

- `.`: any character
- `\w`: any alphanumeric character
- `\s`: any whitespace character
- `\d`: any digit
- Captial variants: \W, \S, \D matches what is NOT alphanumeric, whitespace, and digit respectively.

````python
regexp = r'\w\s\w'

subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'\w\w'

subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'\d\d\d'

subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'\d\d.'

subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'.\d\w'

subject = 'abc 123'

re.search(regexp, subject)
````

### Repeating

- `{}`: repeat a given number of times
- `*`: zero or more
- `+`: one or more
- `?`: optional
- greedy + non-greedy

````python
regexp = r'\w+'
subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'\w+\s\d+'
subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'.+\d'
subject = 'abc 123'

re.search(regexp, subject)
#this considers everything that all the way up until it HAS to consider the next character a digit
````

````python
regexp = r'.+?\d'
subject = 'abc 123'

re.search(regexp, subject)
#adding the ? makes the regexp non greedy, as soon as it has 1 (instead of 1 or more), if the next character is a digit, it will continue and accept it as a digit.
````

````python
regexp = r'\d+'
subject = 'Codeup, founded in 2014, is located at 600 Navarro St. Suite 350, San Antonio, TX 78230. You can find us online at http://codeup.com and our alumni portal is located at https://alumni.codeup.com.'

re.findall(regexp, subject)
````

````python
regexp = r'\d{5}'
subject = 'Codeup, founded in 2014, is located at 600 Navarro St. Suite 350, San Antonio, TX 78230. You can find us online at http://codeup.com and our alumni portal is located at https://alumni.codeup.com.'

re.search(regexp, subject)
````

````python
regexp = r'http.+?\.com'
subject = 'Codeup, founded in 2014, is located at 600 Navarro St. Suite 350, San Antonio, TX 78230. You can find us online at http://codeup.com and our alumni portal is located at https://alumni.codeup.com.'

re.findall(regexp, subject)
````

````python
regexp = r'http.+?\.com'
subject = 'Codeup, founded in 2014, is located at 600 Navarro St. Suite 350, San Antonio, TX 78230. You can find us online at http://codeup.com and our alumni portal is located at https://alumni.codeup.com.'

re.findall(regexp, subject)
````

### Any/None Of:
    - [] anything inside the square brackets
    - [^] anything NOT inside the square brackets

````python
regexp = r'[a1][b2][c3]'
subject = 'abc 123'

re.match(regexp, subject)
````

````python
regexp = r'[^abc]+'
subject = '123abc'

re.search(regexp, subject)
````

````python
regexp = r'[^abc]'
subject = '123abc'

re.findall(regexp, subject)
````

````python
regexp = r'[A-Z][a-z]+'
subject = 'Good afternoon Germain'

re.findall(regexp, subject)
````

````python
regexp = r'\d[2468]'
subject = 'My birthday is on August 14th. My sisters birthday is on February 29th.'

re.findall(regexp, subject)
````

````python
regexp = r'[13579][13579]+'
subject = 'My birthday is on August 14th. My 33 year old sisters birthday is on February 29th.'

re.findall(regexp, subject)
````

````python
regexp = r'\w*[aeiou]\w*'
subject = 'My birthday is on August 14th. My 33 year old sisters birthday is on February 29th.'

re.findall(regexp, subject)
````

### Anchors

- `^` starts with
- `$` ends with
- '\b' word boundary

````python
regexp = r'b'
subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'.+?\b'
subject = 'abc 123'

re.search(regexp, subject)
````

````python
regexp = r'^[aeiouAEIOU]\w+'
subject = 'My birthday is on August 14th. My 33 year old sisters birthday is on February 29th.'

re.findall(regexp, subject)
````

````python
regexp = r'^[A-Z]\w+'
subject = 'My birthday is on August 14th. My 33 year old sisters birthday is on February 29th.'

re.findall(regexp, subject)
````

````python
regexp = r'\w*[A-Z]$'
subject = 'My birthday is on August 14th. My 33 year old sisters birthday is on February 29tH'

re.findall(regexp, subject)
````

````python
regexp = r'^[A-Z]\w*[A-Z]'
subject = 'MY birthday is on August 14th. My 33 year old sisters birthday is on February 29tH'

re.findall(regexp, subject)
````

### Named capture groups

````python
text = 'You should go check out https://regex101.com, it is a great website!'

match = re.search(r'(?P<protocol>https?)://(?P<base_domain>\w+)\.(?P<tld>\w+)', text)
match.groupdict()
````

````python
df.text.str.extract(r'(?P<protocol>https?)://(?P<base_domain>\w+)\.(?P<tld>\w+)')
````

### Verbose regular expressions

- `re.VERBOSE`
- `(?# this is a comment)`

````python
text = 'You should go check out https://regex101.com, it is a great website!'

regexp = r'''
(?P<protocol>https?)
:// (?# ignore the :// that seperates protocol from domain)
(?P<base_domain>\w+)
\.
(?P<tld>\w+)
'''
match = re.search(regexp, text, re.VERBOSE) # whitespace in the regex is ignored
match.groupdict()
````
