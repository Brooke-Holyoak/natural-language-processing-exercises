```python
def parse_news_article(article):
    title = article.h2.text
    date, author = article.select('.italic')[0].find_all('p')
    summary = article.find_all('p')[-1]   
    return {'title': title, 'date': date.text, 'author': author.text, 'summary': summary.text}
    
```

```python
     response = requests.get('https://web-scraping-demo.zgulde.net/news')
     soup = BeautifulSoup(response.text)
     articles = soup.select('.grid.gap-y-12 > div')
     pd.DataFrame([parse_news_article(article) for article in articles])
```

```python
def get_person_info(article):
    name = article.h2.text
    quote = article.find_all('p')[0]
    email = article.find_all('p')[1]
    ph_number = article.find_all('p')[2]
    address = article.find_all('p')[-1]
     
    return {'name': name, 'quote': quote.text.strip(), 'email': email.text, 'ph_number': ph_number.text, 'address': address.text.strip()}
    
```

```python
    response = requests.get('https://web-scraping-demo.zgulde.net/people')
    soup = BeautifulSoup(response.text)
    articles = soup.select('.person')
    #articles = soup.select('div', attrs = {'id':'people'})
    pd.DataFrame([get_person_info(article) for article in articles])
```
